```python
import matplotlib.pyplot as plt

pd.set_option('Display.max_columns', None)
pd.set_option('Display.max_rows', None)
```

```python
# Filter table1 to only have entries >= Year 0
table1_filtered = table1.reset_index()
table1_filtered = table1_filtered[table1_filtered.Year > 0]

table1_filtered.head()
```

```python
table1_filtered.info()
```

```python
def run_simulation(system, df):
    results = [0]
    
    for r in df_HYDE.values:
        net_growth = (system.alpha * r[1]) + (system.beta * r[1])
        results.append(net_growth)

    return results
```

```python
#Run the model using only one population estimate (column) from table1_filtered
#Per the table1.info results, HYDE has the most values (29 out of 46)

#Create the system object
system = System(alpha=0.025,
                beta=0.018)

#Create the HYDE table and remove NaN's (nulls)
df_HYDE = table1_filtered[['Year', 'HYDE']]
df_HYDE = df_HYDE.dropna(axis=0, how='any')
df_HYDE.head(1)

#Run the simulation and add results as a column in df_HYDE
results = run_simulation(system, df_HYDE)
results = results[1:]

df_HYDE['forecast'] = results

plt.plot(df_HYDE.Year, df_HYDE.HYDE, color='blue')
plt.plot(df_HYDE.Year, df_HYDE.forecast, color = 'red')
plt.show()
```
